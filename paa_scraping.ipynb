```python
from collections import defaultdict
from bs4 import BeautifulSoup
import requests

with open("url_professores.txt", "r") as arquivo:
        urls = arquivo.read()

all_links = urls.split("\n")[:-1]

todos_professores = []

for url in all_links:
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')

    professor = []

    professor.append(soup.find('span', {'class' : 'contact-name'}).get_text())

    tags = ",".join([x for x in ''.join(soup.find('div', {'class' : 'tags'}).get_text()).replace("\t", "%").replace("\n", "%").split("%") if len(x) > 0][1:])

    professor.append(tags)

    informacoes = soup.find('span', {'class' : 'contact-misc'}).get_text().replace("\n", " ")

    while(not informacoes[0].isalnum()):
        informacoes = informacoes[1:]

    professor.append(informacoes)

    todos_professores.append(professor)
```

```python
sigaa_info = defaultdict(list)

last_disciplina = ""

for link in soup.find_all(['td', 'span'], {'class': ['nome', 'tituloDisciplina']}):
  if link.text[-2:] != 'h)':
    last_disciplina = ' '.join(link.text.split(" ")[2:])
  else:
    sigaa_info[last_disciplina].append(' '.join(link.text.split(" ")[:-1]))

for i in sigaa_info.keys():
  sigaa_info[i] = list(Counter(sigaa_info[i]).keys())
```

```python
final_sigaa = defaultdict(list)

# PIVOTEAMENTO DO DICIONARIO
for i in sigaa_info.keys():
  for o in sigaa_info[i]:
    z = o.split(" ")
    final_sigaa[z[0] + " " + z[-1]].append(i)
```
